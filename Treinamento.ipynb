# Treinamento do Modelo

* **Img 640**: Define a dimensão da imagem de entrada como 640x640 pixels.
* **Epochs 100**: Especifica o número de épocas para o treinamento, ou seja, o número de vezes que o modelo será treinado em todo o conjunto de dados, neste caso eu defini como 100.
* **Batch-size 8**: Define o tamanho do lote para o treinamento, ou seja, o número de amostras processadas antes de atualizar os pesos do modelo.
